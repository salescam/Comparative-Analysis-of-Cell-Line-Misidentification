```python
import scipy.stats as stats
import numpy as np
contaminant_cell_type_data = [actual_cell_type_count_merged]
cervical_adenocarcinoma_count = 140
#average_contaminated_per_cell_type is the mean of the contaminations per cell type
data_array = np.array(actual_cell_type_count_merged)
t_statistic, p_value = stats.ttest_1samp(data_array, cervical_adenocarcinoma_count)
print("The mean of contaminations per cell type is:", average_contaminated_per_cell_type)
print("t-statistic:", t_statistic)
print("p-value:", p_value)
```
